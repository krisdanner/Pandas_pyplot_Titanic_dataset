```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_profiling as pp
import plotly.graph_objs as go
from plotly.offline import iplot
import plotly.express as px


df = pd.read_csv("https://raw.githubusercontent.com/Gabrielmarcial/Data-Marte/main/datasets/train.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe().T
```

```python
f, ax = plt.subplots(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, linewidths=.5, fmt='.1f', ax=ax)
plt.show()
```

```python
mulheres = df.loc[df.Sex == 'female']['Survived']
taxa_mulheres = round(sum(mulheres)*100/len(mulheres), 2)
print('Mulheres sobreviventes: ', taxa_mulheres, '%')
```

```python
homens = df.loc[df.Sex == 'male']['Survived']
taxa_homens = round(sum(homens)*100/len(homens), 2)
print('Homens sobreviventes: ', taxa_homens, '%')
```

```python
df.Age.min()
```

```python
criancas1ano = df.loc[df.Age == 1.0]
print(len(criancas1ano))
```

```python
plt.style.available
plt.style.use('seaborn-colorblind')
plt.rcParams['figure.figsize'] = (10,7)
plt.title('Idades dos Passageiros do Titanic', size=15)
plt.xlabel('Passageiro', size=15)
plt.ylabel('Idade', size=15)
plt.plot(df.Age, '.-g')
plt.show()
```
